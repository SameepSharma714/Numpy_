```python
#Deep and shallow copy
#deep copy
import copy

l1 = [1, [2, 3]]
l2 = copy.deepcopy(l1)   # deep copy

l2[1][0] = 99

print(l1)   # [1, [2, 3]]   ← original NOT changed
print(l2)   # [1, [99, 3]]
```

```python
#shallow copy
l1 = [1, [2, 3]]
l2 = l1.copy()   # shallow copy

l2[1][0] = 99
print(l1)  # [1, [99, 3]]  ← original changed
print(l2)  # [1, [99, 3]]
```
